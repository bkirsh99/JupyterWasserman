# [Steps to write and execute a script](http://omgenomics.com/writing-bash-script/):
> - "nano **myscript**" opens the file editor to create/edit the script "myscript"  
> - "chmod +x **myscript**" makes the script executable  
> - "./**myscript**" or, more generally, "*path_to_script*/**myscript**" runs the script
